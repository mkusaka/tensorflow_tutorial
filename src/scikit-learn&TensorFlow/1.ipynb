# 機械学習の基礎
- OCRなどで昔から存在
- 数億人規模への貢献という視点からすると最初のMLの貢献はスパムフィルタが最初と思われる

## 1.1 機械学習とはなにか
- `コンピュータがデータから学習できるようにするためのコンピュータプログラミングについての科学`
- 広義には `明示的にプログラミングせずコンピュータに学習能力を与えるための学習分野`
- 技術的に経験をEとし、タスクをTとすると `EによってタスクTに対する測定指標Pで測定した性能が上がるとき` がTについてEから学習したと言われる

### スパムメールの例
- スパムメールと正常なメール(ハムと呼ぶ)を与えるとスパムフィルタがスパムを見分けられるようにになるならフィルタは機械学習プログラムである
- スパムフィルタのタスクT: 新しいメールにスパムのフラグを付けるかどうかを判定すること
- 性能指標Pは正しく分類された割合など
- Wikipediaのコピーをダウンロードするだけではただデータが不植えただけなので機械学習ではない


**def: 訓練セット**
```
システムが学習のために使うデータ例
```

**def: 訓練インスタンス**
```
訓練セットの個々のデータ例
**標本**、  **サンプル** とも呼ぶ
```

**def: 訓練データ**
```
- スパムのフラグを付けるためのデータ
- 経験Eにあたる
```

#### def: 精度(accuracy)
性能指標Pのこと(ここでは正しく分類された割合など)

## 1.2 なぜ機械学習を使うのか
従来からのプログラミングテクニックでスパムフィルタを作ると辛い

### 従来の方法でスパムフィルタを作った場合
1. 一般的な特徴に着目する
2. 気づいたパターンごとに検出アルゴリズムを各
3. プログラムをテストし、十分良い結果が得られるまで1. 2. を繰り返す

#### 問題点
問題が単純ではないので複雑なルールの長いリストを含んだものになる→メンテが大変
- スパムフィルタにスパマーが気づいて対策した場合も追加しなければならない→スパマーとのいたちごっこのルール追加をずっとしなければならない

### 機械学習テクニックを使ったスパムフィルタ
ハムのデータ例とスパムをを比べて異常に多く見られる語句を検出し、スパムを見分けるために使える語句を自動的に学習
- プログラムはずっと短くなり、メンテしやすくなり、正確にもなる
- スパムフィルタの更新も自動で行うので、人間が特に何かをしなくてもフラグを付けるようになる(もちろん自動でフラグを付けるようにする実装は必要そうだが。。。)


### 機械学習が力を発揮する場所
従来のアプローチでは複雑になりすぎる問題や、既知のアルゴリズムがわかっていない問題に対して力を発揮する
- 音声認識など
   - 頑張ってハードコーディングすると、ノイズもあり数百、数千万の異なる人の言語を見分けるところまでスケールアップは難しい
   - 個々の単語の発音例の録音を与えて自動で学習するアルゴリズムを書くことが大事→機械学習

**def: データマイニング**
```
大量のデータをMLで掘り下げてすぐにはわからなかったパターンを見つけること
```

### まとめ
- 既存のソリューションでは手作業が大量に発生する or ルールの長いリストが必要な場合
- 伝統的な方法で良いソリューションが得られない複雑な問題
- 変動する環境
- 複雑なシステムや大量のデータについての知見の獲得

## 1.3 機械学習システムのタイプ
分類パターン(それぞれ相互排他的ではなく、組み合わせて使う)
- 人間関与のもと訓練されるか否か
  - 教師あり
  - 教師なし
  - 半教師あり
  - 強化学習
- その場で少しずつ学習できるかどうか
  - オンライン学習
  - バッチ学習
- 新しいデータポイントと既知のデータポイントを比較する or 訓練データからパターンを見つけ出して予測モデルを構築するか
  - インスタンスベース学習
  - モデルベース学習

### 1.3.1 教師あり/なし学習
#### 1.3.1.1 教師あり学習
**def: 教師あり学習**
```
アルゴリズムに与える訓練データの中にラベルと呼ばれる答えが含まれている
```

##### タスク例
- 分類
  - クラス(スパムかハムか)が明示されたメールの例を大量に与えて訓練する

**def: 回帰**
```
予測子(一連の特徴量)からターゲットの数値を予測すること
```

#### 1.3.1.2 教師なし学習
**def: 教師なし学習**
```
ラベルが付いていない訓練データをもとに学習する形式
```

#### 1.3.1.3 半教師あり学習
**def: 半教師あり学習**
```
僅かなラベルが付いたデータとそれ以外のデータをもとに学習する形式
```
##### 例
- Google photoなど

#### 1.3.1.4 強化学習
**def: 強化学習**
```
環境を観察し、行動を選択して実行し、報酬と呼ばれるものを得るように学習する形式
```

**def: エージェント**
```
強化学習において実際に学習するシステム
```

**def: 方策**
```
強化学習においてエージェントが学習する戦略、報酬が高くなるように学習していく
```

### 1.3.2 バッチ学習とオンライン学習
#### 1.3.2.1 バッチ学習
**def: バッチ学習**
```
すべての訓練データを事前に与え、それ以上は学習させずに実行し続ける形式
オフライン学習とも呼ばれる
```

**lem: バッチ学習のデメリット**
```
- 新しいデータについての知識を与えたいときは完全なデータセットをもとに0からが憂愁させる必要がある
- 新しいシステムに移動する際は古いシステムを止めて完全移行する必要がある
- すべてのデータを用いた学習は時間がかかるため、必然と更新に時間もかかる
- リソースを大量に食うため、コストも掛かる
```
※とはいえ学習の自動化は簡単(ほんまか？)なので変化に対応するのは簡単(ほんまか？)

**lem: バッチ学習のメリット**
```
- システが単純
- うまく動作することが多い
```

#### 1.3.2.2 オンライン学習
**def: オンライン学習**
```
小さなグループごとに地区図手時に学習させる方式
```

**lem: オンライン学習のメリット**
```
- 差分更新なので高速かつコストが掛からない
- 更新分の差分はすぐに学習が済ませることができる
- 大きなメモリを必要としないので、莫大なデータなど通常のメモリに乗らないデータも処理できる
```
